# scatter plot
### to check the correlation of 2 or 3 series in a dataset

```python
irisdf.plot(kind='scatter',x ='petal_length',y='petal_width')
plt.show()
```

```python
irisdf.plot(kind='scatter',x ='petal_length',y='petal_width',
            c='petal_length',cmap='cool',marker='+',alpha=.5,s=200,figsize=(15,5))
plt.xticks([])
plt.show()
```
